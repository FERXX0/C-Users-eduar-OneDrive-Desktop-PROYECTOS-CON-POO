```python
import random
class Triqui():
  def __init__(self):
      self.tablero = []
        
  def crear_tablero(self):
      for i in range(3):
          fila = []
          for j in range(3):
              fila.append('-')
          self.tablero.append(fila)

  def primer_jugador_aleatorio(self):
      return random.randint(0, 1)

  def arreglar_la_cuadricula(self, fila, columna, jugador):
      self.tablero[fila][columna] = jugador

  def jugador_ganador(self, jugador):
      ganar = None

      n = len(self.tablero)

        
      for i in range(n):
          ganar = True 
          for j in range(n):
              if self.tablero[i][j] != jugador:
                  ganar = False
                  break
          if ganar:
              return ganar

        
      for i in range(n):
          ganar = True
          for j in range(n):
              if self.tablero[j][i] != jugador:
                  ganar = False
                  break
          if ganar:
              return ganar

  
      ganar = True
      for i in range(n):
          if self.tablero[i][i] != jugador:
              ganar = False
              break
      if ganar:
          return ganar

      ganar = True
      for i in range(n):
          if self.tablero[i][n - 1 - i] != jugador:
              ganar = False
              break
      if ganar:
          return ganar
      return False

      for fila in self.tablero:
          for articulo in fila:
              if articulo == '-':
                  return False
      return True

  def tablero_lleno(self):
      for fila in self.tablero:
          for articulo in fila:
              if articulo == '-':
                  return False
      return True

  def turno_del_otro_jugador(self, jugador):
      return 'X' if jugador == 'O' else 'O'

  def mostrar_tablero(self):
      for fila in self.tablero:
          for articulo in fila:
              print(articulo, end=" ")
          print()

  def comenzar(self):
      self.crear_tablero()

      jugador = 'X' if self.primer_jugador_aleatorio() == 1 else 'O'
      while True:
          print(f"Jugador {jugador} es tu turno")

          self.mostrar_tablero()

          fila, columna = list(
              map(int, input("Introduce el numero de la fila y la columna con espacios y sin comas: ").split()))
          print()

          self.arreglar_la_cuadricula(fila - 1, columna - 1, jugador)

          if self.jugador_ganador(jugador):
              print(f"!El jugador {jugador} gano la partida¡")
              break

          if self.tablero_lleno():
              print("!La partida quedo en empate¡")
              break

          jugador = self.turno_del_otro_jugador(jugador)

      print()
      self.mostrar_tablero()

objeto = Triqui()
objeto.comenzar()
```
