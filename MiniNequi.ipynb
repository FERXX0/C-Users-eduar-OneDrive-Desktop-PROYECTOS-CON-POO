```python
#MiniNequiconPOO
class Usuarios:
  def __init__(self):
    self.Nombre=[]
    self.claves=[]
    self.dinero=[]
    self.dinero_2=[]

    for i in range(2):
      self.dinero.append(0)
      self.dinero_2.append(0)
      self.Nombre.append(input("Inscriba su nombre: "))
      self.claves.append(input("Inscriba su contraseña: "))
      print("Su cuenta es la N°",i)

class Nequi(Usuarios):
  def Entrada(self):
    self.posicion=int(input("Escriba su N° de cuenta: "))
    nombre=input("Escriba su nombre: ")
    claves=input("Escriba su contraseña: ")
    if nombre==self.Nombre[self.posicion] and claves==self.claves[self.posicion]:
       Iniciar.inicio()
    else:
      print("❌El Nombre o la Contraseña son incorrectos❌")

  def inicio(self):
    print("Hola👋🏻",self.Nombre[self.posicion], "Su saldo es de: $", self.dinero[self.posicion],"💸")
    print("Tiene un ahorro de: $",self.dinero_2[self.posicion],"💸")
    print("1.Cuenta ahorro")
    print("2.Depositos")
    print("3.Retiros")
    print("4.Cambiar De Cuenta")
    print("5.Exit")

    self.opcion=input("Escoja una Opción: ")
    if self.opcion=="1":
      Iniciar.ahorrando()
      Iniciar.inicio()
    elif self.opcion=="2":
      Iniciar.depositando()
      Iniciar.inicio()
    elif  self.opcion=="3":
      Iniciar.retirando()
      Iniciar.inicio()
    elif self.opcion=="4":
      Iniciar.Entrada()
    elif self.opcion=="5":
      exit()

  def depositando(self):
    depositos=int(input("Deposita la cantidad desea: "))
    depositos=self.dinero[self.posicion]+depositos
    self.dinero.pop(self.posicion)
    self.dinero.insert(self.posicion,depositos)


  def retirando(self):
    retirar=int(input("El valor a retirar es de: "))
    if  self.dinero[self.posicion]>=retirar:
      retira=self.dinero[self.posicion]-retirar
      self.dinero.pop(self.posicion)
      self.dinero.insert(self.posicion,retirar)
    else:
      print("KBRÓN, NO TIENES ESE DINERO😡")

  def ahorrando(self):
      ahorro_dinero_2=int(input("La cantidad a depositar en la cuenta de ahorros es de: "))
      if  self.dinero[self.posicion]>=ahorro_dinero_2:
        ahorro_dinero_2_entra=self.dinero_2[self.posicion]+ahorro_dinero_2
        self.dinero_2.pop(self.posicion)
        self.dinero_2.insert(self.posicion,ahorro_dinero_2_entra)
        ahorro_dinero_2_sale=self.dinero[self.posicion]-ahorro_dinero_2
        self.dinero.pop(self.posicion)
        self.dinero.insert(self.posicion,ahorro_dinero_2_sale)
      else:
          print("RASCABICHO, QUIERES AHORRAR Y DEPOSITAS UNA MISERIA🤡")


Iniciar=Nequi()
Iniciar.Entrada()
```
